```python
import globals as gl
import os
import numpy as np
import PcmPy as pcm
from matplotlib import pyplot as plt

plt.style.use('default')
```

```python
H = 'L'
roi = 'M1'

sn = 104

npz = np.load(os.path.join(gl.baseDir, gl.pcmDir, f'G_obs.{sn}.{H}.{roi}.npz'))
G = npz['G_obs']
conds = npz['conds']
D = pcm.G_to_dist(G)

fig, ax = plt.subplots()

ax.imshow(D, vmin=-.01, vmax=.15)
ax.set_xticks(np.arange(len(conds)))
ax.set_xticklabels(conds, rotation=90)
ax.set_yticks(ax.get_xticks())
ax.set_yticklabels(conds)

cax = ax.get_images()[0]
cbar = fig.colorbar(cax, ax=ax, orientation='vertical', fraction=.01)
cbar.set_label('dissimilarity (a.u.)')

ax.set_title(roi)

plt.show()
```
