```python
import globals as gl
from main import main
import matplotlib.pyplot as plt
```

```python
experiment = 'efc2'
participant_id = 'subj100'
session = 'testing'
day = '1'
ntrial = 31

fig, axs = plt.subplots(1, 2, sharey=True)
_, ax, chordID, day, chord = main('PLOT:force_in_trial', experiment, participant_id, session, day, ntrial, fig=fig, axs=axs[0])
ax.set_ylabel('force (N)')
ax.set_title(f'{chordID}, day{day}, {chord}')

day = '5'
ntrial = 37
_, ax, chordID, day, chord = main('PLOT:force_in_trial', experiment, participant_id, session, day, ntrial, fig=fig, axs=axs[1])
ax.set_title(f'{chordID}, day{day}, {chord}')

fig.legend(gl.channels['force'], loc='upper center', ncol=5)

fig.supxlabel('time (s)')
```
