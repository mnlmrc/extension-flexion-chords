```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os
import globals as gl
import numpy as np

plt.style.use('default')

experiment = 'EFC_learningEMG'
```

```python
df = pd.read_csv(os.path.join(gl.baseDir, experiment, 'coherence', 'phase_consistency.tsv'), sep='\t')
df = df[df['trialPoint']==1]
df = df.groupby(['sn', 'chord', 'day', 'frequency']).mean(numeric_only=True).reset_index()

fig, axs = plt.subplots(1, 2, sharex=True, sharey=True)

for d, day in enumerate([1, 5]):
    ax = axs[d]
    sb.lineplot(data=df[df['day']==day], ax=ax, x='frequency', y='plv_raw', hue='chord', errorbar='se', err_kws={'lw': 0}, palette=['red', 'blue'], hue_order=['trained', 'untrained'], legend=True if day == 5 else False)
```

```python
df = pd.read_csv(os.path.join(gl.baseDir, experiment, 'coherence', 'phase_consistency.tsv'), sep='\t')
df = df[df['trialPoint']==1]
df = df.groupby(['sn', 'chord', 'day', 'frequency']).mean(numeric_only=True).reset_index()

fig, axs = plt.subplots(1, 2, sharex=True, sharey=True)

for d, day in enumerate([1, 5]):
    ax = axs[d]
    sb.lineplot(data=df[df['day']==day], ax=ax, x='frequency', y='plv0', hue='chord', errorbar='se', err_kws={'lw': 0}, palette=['red', 'blue'], hue_order=['trained', 'untrained'], legend=True if day == 5 else False)
    ax.set_xscale('log')
    ax.axvline(2, lw=.8, ls='--', color='k')
    ax.axvline(10, lw=.8, ls='--', color='k')
    ax.axvline(50, lw=.8, ls='--', color='k')
```
